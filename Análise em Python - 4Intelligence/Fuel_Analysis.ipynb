### 1.a. Qual a Sazonalidade da venda de combustíveis?

```python
# Gasolina

gasolina_melted = pd.melt(gasolina,id_vars=['regiao','meses'],var_name='Ano',value_name='Vendas')

gasolina_melted.head()
```

```python
# Br
sns.set_style('darkgrid')

gasolina_melted_br = gasolina_melted[gasolina_melted['regiao']=='br']

plt.figure(figsize=(15,5))

sns.lineplot(data=gasolina_melted_br,x='meses',y='Vendas',hue='Ano',palette='colorblind')
plt.title('Vendas BR de gasolina - Mês a mês - 2000 a 2020')
plt.ylabel('Vendas em m³ de Gasolina')
plt.legend(title='Ano',loc='lower left',ncol=3)

plt.show()
```

##### Sazonalidade Gasolina
    
        Vemos aqui que, para a Gasolina, o Desempenho BR Mês a Mês, segundo a visualização acima, parece tem uma tendência a aumento de vendas no mês de Dezembro.
        Isto provavelmente se dá por conta das diversas viagens de ano novo e Natal, que fazem parte das tradições das famílias aqui do Brasil

```python
# Testando, também, Etanol e Diesel

# Fazendo o melt para ambos

etanol_melted = pd.melt(etanol,id_vars=['regiao','meses'],var_name='Ano',value_name='Vendas')
diesel_melted = pd.melt(diesel,id_vars=['regiao','meses'],var_name='Ano',value_name='Vendas')
```

```python
# Gráfico do Etanol
sns.set_style('darkgrid')

etanol_melted_br = etanol_melted[etanol_melted['regiao']=='br']

plt.figure(figsize=(15,5))

sns.lineplot(data=etanol_melted_br,x='meses',y='Vendas',hue='Ano',palette='colorblind')
plt.title('Vendas BR de etanol - Mês a mês - 2000 a 2020')
plt.ylabel('Vendas em m³ de Etanol')
plt.legend(title='Ano',loc='lower left',ncol=3)

plt.show()

# Gráfico do Diesel

diesel_melted_br = diesel_melted[diesel_melted['regiao']=='br']

plt.figure(figsize=(15,5))

sns.lineplot(data=diesel_melted_br,x='meses',y='Vendas',hue='Ano',palette='colorblind')
plt.title('Vendas BR de diesel - Mês a mês - 2000 a 2020')
plt.ylabel('Vendas em m³ de Diesel')
plt.legend(title='Ano',loc='lower left',ncol=3)

plt.show()
```

##### Sazonalidade Etanol

        Para o Etanol vemos, mesmo que de forma mais tímida, uma repetição do padrão estabelecido pela gasolina: O consumo deste combustível passa por uma alta nos períodos próximos às festas de fim de ano, muito por conta dos comportamentos costumeiros das famílias brasileiras. É claro que seu resultado é mais tímido, dada a menor adoção dessa categoria em alguns estados.

##### Sazonalidade Diesel

        Para o Diesel, por outro lado, a história é diferente. Ao contrário dos outros dois, mas muito provavelmente também por conta das festas de fim de ano, o pico ocorre próximo do mês 10, Outubro.
        Isto provavelmente ocorre porque o Diesel é o combustível utilizado pela frota de caminhões brasileira, responsável pela maior parte do carregamento de produtos para cidades, produtos estes que passam a ser estocados nas lojas para o fim do ano.
        Dessa forma, com maior demanda das lojas, ocorrem, consequentemente, mais viagens de caminhão e, portanto, maior uso do combustível Diesel.

### 1.b. Essa Sazonalidade é igual em todos os estados?

```python
# Para visualizações comparativas entre estados, utilizarei os últimos 10 anos: 2020, 2019, 2018, 2017, 2016
anos = [2010,2011,2012,2013,2014,'2016',2017,2018,2019,2020] # Por algum motivo somente 2016 era uma string

gasolina_10yr = gasolina[['regiao','meses']+anos]
etanol_10yr = etanol[['regiao','meses']+anos]
diesel_10yr = diesel[['regiao','meses']+anos]
```

```python
# Gasolina
gasolina_10yr_melted = pd.melt(gasolina_10yr,id_vars=['regiao','meses'],var_name='Ano',value_name='Vendas')

gasolina_10yr_melted.head()
```

```python
gasolina_10yr_melted_estados = gasolina_10yr_melted[gasolina_10yr_melted['regiao']!='br']
```

```python
gasolina_10yr_melted_estados_norm = gasolina_10yr_melted_estados.copy()

# Normalização para facilitar visualização
gasolina_10yr_melted_estados_norm['Vendas_norm'] = gasolina_10yr_melted_estados.groupby('regiao')['Vendas'].transform(
lambda x: (x-x.min())/(x.max()-x.min())
)
```

```python
sns.set_style('darkgrid')

sns.relplot(kind='line',data=gasolina_10yr_melted_estados_norm,x='meses',y='Vendas_norm',hue='Ano',col='regiao',col_wrap=4)\
.set_axis_labels('Meses','Vendas normalizadas')

plt.show()
```

##### Sazonalidade da Gasolina por Estado disponível
    
        Como podemos ver, o crescimento da demandas por gasolina parece, em geral, se reproduzir em todos os estados. Por outros lado, é evidente que diferentes estados possuem outros picos de demandas em diferentes momentos do ano.
        Em geral, quando segmentado por estado, é possível ver que há muito mais variabilidade nos picos de demanda, muito provavelmente atrelada a especificidades locais. No entanto, apesar dessas variações que fazem com que a sazonalidade não seja exatamente igual entre os estados, acredito que é seguro dizer que deve-se, pelo menos, esperar uma alta demanda por Gasolina nas épocas de fim de ano na maior parte dos estados.

```python
# Etanol
etanol_10yr_melted = pd.melt(etanol_10yr,id_vars=['regiao','meses'],var_name='Ano',value_name='Vendas')
etanol_10yr_melted_estados = etanol_10yr_melted[etanol_10yr_melted['regiao']!='br']
```

Enquanto em SP, local com a maior demanda por combustível em geral e MG, o segundo colocado, e até mesmo MT, vemos que a Sazonalidade de fim de ano, de alguma forma, continua a se reproduzir, especialmente em anos mais recentes.

### 1.c. Distribuição do consumo de Combustíveis ao entre os estados brasileiros

```python
# Convertendo 2016 de String para Numérico.
gasolina_melted['Ano']=pd.to_numeric(gasolina_melted['Ano'])
etanol_melted['Ano']=pd.to_numeric(etanol_melted['Ano'])
diesel_melted['Ano']=pd.to_numeric(diesel_melted['Ano'])

anos_integers = [2010,2011,2012,2013,2014,2016,2017,2018,2019,2020]
```

```python
# Gasolina
gasolina_melted_estados = gasolina_melted[gasolina_melted['regiao']!='br']
```

```python
gasolina_melted_estados_5yrs = gasolina_melted_estados[gasolina_melted_estados['Ano'].isin(anos_integers)]
gasolina_melted_estados_5yrs.head()
```

```python
# Pivotando o gasolina_melted_estados_5yrs, melhor maneira de fazer um stack bar plot

gasolina_pivot_estados_5yrs = gasolina_melted_estados_5yrs.pivot(index=['meses','Ano'], columns='regiao',values='Vendas').reset_index()
gasolina_pivot_grouped_5yrs = gasolina_pivot_estados_5yrs.groupby('Ano').sum().reset_index()
```

```python
gasolina_pivot_grouped_5yrs_sorted = gasolina_pivot_grouped_5yrs[['Ano','sp','mg','go','df','pa','ma','mt','to']]
gasolina_pivot_grouped_5yrs_sorted.head(2)
```

```python
sns.set_style('darkgrid')

plot_stackbar = gasolina_pivot_grouped_5yrs_sorted.set_index('Ano').plot(kind='bar',stacked=True,figsize=(10,5))
plt.xticks(rotation=0)
plt.legend(title='Região',loc='lower left',bbox_to_anchor=(1,0))
    
plt.show()
```

```python
gasolina_pivot_grouped_5yrs_sorted_total = gasolina_pivot_grouped_5yrs_sorted.copy()

gasolina_pivot_grouped_5yrs_sorted_total['total'] = gasolina_pivot_grouped_5yrs_sorted\
[['sp','mg','go','df','pa','ma','mt','to']].sum(axis=1)
```

```python
gasolina_pivot_grouped_5yrs_sorted_total.head(2)
```

```python
gasolina_pivot_grouped_5yrs_sorted_norm = gasolina_pivot_grouped_5yrs_sorted.copy()

# Normalização
col_ano = ['Ano']

for col in gasolina_pivot_grouped_5yrs_sorted_norm.columns:
    if col not in col_ano:
        gasolina_pivot_grouped_5yrs_sorted_norm[col] = (gasolina_pivot_grouped_5yrs_sorted_total[col]/gasolina_pivot_grouped_5yrs_sorted_total['total'])*100
    
```

```python
gasolina_pivot_grouped_5yrs_sorted_norm.head(2)
```

```python
sns.set_style('darkgrid')

plot_stackbar = gasolina_pivot_grouped_5yrs_sorted_norm.set_index('Ano').plot(kind='bar',stacked=True,figsize=(10,5))
plt.xticks(rotation=0)
plt.legend(title='Região',loc='lower left',bbox_to_anchor=(1,0))

for c in plot_stackbar.containers:
    plot_stackbar.bar_label(c,fmt='%.2f%%',label_type='center',size=8)

plt.show()
```

### 1.d. Evolução do mercado nos útimos 5 anos

```python
last_5_years = [2016,2017,2018,2019,2020]
```

```python
gasolina_melted_br_5yr = gasolina_melted_br.copy()
etanol_melted_br_5yr = etanol_melted_br.copy()
diesel_melted_br_5yr = diesel_melted_br.copy()
```

```python
gasolina_melted_br_5yr['Ano'] = pd.to_numeric(gasolina_melted_br_5yr['Ano'])
etanol_melted_br_5yr['Ano'] = pd.to_numeric(etanol_melted_br_5yr['Ano'])
diesel_melted_br_5yr['Ano'] = pd.to_numeric(diesel_melted_br_5yr['Ano'])
```

```python
gasolina_melted_br_5yr = gasolina_melted_br_5yr[gasolina_melted_br_5yr['Ano'].isin(last_5_years)]
etanol_melted_br_5yr = etanol_melted_br_5yr[etanol_melted_br_5yr['Ano'].isin(last_5_years)]
diesel_melted_br_5yr = diesel_melted_br_5yr[diesel_melted_br_5yr['Ano'].isin(last_5_years)]
```

```python
# Criando sinalizadores e unindos os dfs
gasolina_melted_br_5yr['kind'] = 'Gasolina'
etanol_melted_br_5yr['kind'] = 'Etanol'
diesel_melted_br_5yr['kind'] = 'Diesel'


union_5yr = pd.concat([gasolina_melted_br_5yr,etanol_melted_br_5yr,diesel_melted_br_5yr])

union_5yr['kind'].unique()
```

```python
union_5yr.head()
```

```python
union_5yr_grouped = union_5yr.groupby(['Ano','meses'])['Vendas'].sum().reset_index()

union_5yr_grouped.head()
```

```python
plt.figure(figsize=(12,5))

sns.lineplot(data=union_5yr,x='Ano',y='Vendas',hue='kind',marker='o',errorbar=None)

plt.title('Desempenho por Combustível nos último 5 anos')
plt.legend(title='Combustível',loc='upper left',bbox_to_anchor=(1,1))

plt.show()
```

##### Como evoluiu nos últimos 5 anos?

        O de Diesel passou por um leve crescimento se comparado, em questão de variação, com o do Etanol. Falando disso, o Etanol comparativamente com a gasolina parece ter ganhado mais participação no mercado, enquanto a gasolina tem perdido.

```python
plt.figure(figsize=(12,4))

sns.lineplot(data=union_5yr_grouped,x='Ano',y='Vendas',marker='o',errorbar=None)

plt.title('Desempenho por Combustível nos último 5 anos')

plt.show()
```

##### Qual o tamanho do mercado de combustíveis hoje em dia?

        No ano de 2020 o mercado parece grande o suficiente para que distribuir um total aproximado de 9.4 milhões de metros cúbicos totais de combustível. Podemos ver, porém, que em 2019 este mercado foi muito maior do que é hoje em sua última aferição no ano de 2020.
